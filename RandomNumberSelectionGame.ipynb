```python
i = 1
file = open("number_file.txt", "w+")
print("Enter 50 integer values :")
while i <= 10:
    inputed_value = int(input())
    file.write(str(inputed_value))
    file.write("\n")
    i = i + 1
file.close()
```

```python
import random

total_scores = 0
high_scores = 0

choice = 'yes'
while choice == 'yes':
    data_list = []
    data_list = open("data_file.txt").read().splitlines()
    data_grid_view = []

    grid_size = 0
    while grid_size <= 25:
        random_number = random.randint(0,37)
        if data_list[random_number] not in data_grid_view:
            data_grid_view.append(data_list[random_number])
            grid_size += 1 
    
    print(*data_grid_view[0:5],sep ='   ')
    print('\n')
    print(*data_grid_view[5:10],sep ='   ')
    print('\n')
    print(*data_grid_view[10:15],sep ='   ')
    print('\n')
    print(*data_grid_view[15:20],sep ='   ')
    print('\n')
    print(*data_grid_view[20:25],sep ='   ')
    print('\n')
    
    
    selected_number = random.choice(data_grid_view)
    selected_number = int(selected_number)
    print(selected_number)
    
    total_guesses = 1
    guesses_left = 6
    
    while total_guesses <= 5 :
        guesses_taken = guesses_left - total_guesses
        print("\nYou have ", guesses_taken, "Guesses Left")
        user_selected = int(input("Choose a Number : "))
        
        if user_selected > selected_number:
            print("\nYou Choose Greater Number :( ")
        elif user_selected < selected_number:
            print("\nYou Choose Lesser Number :( ")
        elif user_selected == selected_number:
            break
        if user_selected not in data_grid_view:
            print("\nYou Choose Number that is out of Data Grid View :( ")
            guesses_taken = guesses_left + total_guesses
            
        total_guesses = total_guesses + 1
    
    if user_selected == selected_number:
        print("\nYou Choose correct number in ", total_guesses, "Guesses :) \n")
        total_scores = total_scores + 10
    
    elif user_selected != selected_number:
        total_scores = total_scores - 2
        
        print("\nYou Failed in all tries :( ")

    
    if total_scores > high_scores:
        high_scores = total_scores
        
    choice = input("\nare you want to continue Game again? yes/no    \n" )
    if choice == 'no':
        break
    
print("\nTotal Scores are : ", total_scores)
print("\nHigh Scores are : ", high_scores)
print("\nGame Over")
```
